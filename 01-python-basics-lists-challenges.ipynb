```python
my_list = [3, 1, 4, 1, 5, 9]

print(my_list[-1])   # 9
print(my_list[-2])   # 5
print(my_list[-3])   # 1
print(my_list[-6])   # 3   ← same as [0]

# Change length and see -1 still works
my_list.append(100)
print("After append, last element:", my_list[-8])   # still 100, no need to change code
```

```python
### Key Takeaways on Negative Indexing
- `-1` always means **last element** (same as `[len(my_list)-1]`)
- `-k` means **k-th element from the end**
- Valid range: from `-len(my_list)` to `-1`
- Going beyond (e.g. `-len-1`) → IndexError, same as positive out-of-range
- Advantage: length-independent, readable, Pythonic
```

```python
# Original list
print("Original last:", my_list[-1])           # 9

# Append two elements
my_list += [77, 88]
print("After append, last is still -1:", my_list[-1])   # should be 88
print("Second-to-last:", my_list[-2])                   # should be 77
```

```python
### Safe access for empty lists
```

```python
empty = []

# This will crash — don't run it uncommented
# print(empty[0])                # IndexError

# Safe pattern 1: explicit check
if empty:
    print("First:", empty[0])
    print("Last:", empty[-1])
else:
    print("List is empty — no elements to access")

# Safe pattern 2: ternary (very common in Python)
first_safe = empty[0] if empty else "No first element"
last_safe  = empty[-1] if empty else "No last element"
print("Safe first:", first_safe)
print("Safe last :", last_safe)

# Bonus: Pythonic short version (often seen in data code)
print("Safe first (short):", empty[0] if empty else None)
```
